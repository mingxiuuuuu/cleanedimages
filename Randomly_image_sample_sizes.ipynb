Sample 100 images

```
import shutil, random, os 
dirpath = '/content/cleanedimages/Cleanedimages'
destDir = '/content/dresses100'

filenames = random.sample(os.listdir(dirpath),100)

for fname in filenames:
  srcpath = os.path.join(dirpath,fname)
  shutil.copy(srcpath,destDir)
```

Download into load machine, we will upload into github repository later

```
from google.colab import files
#For the folder you have to zip it first and can only download later on
!zip -r dresses100.zip /content/dresses100
#Download files
files.download('dresses100.zip')
```
